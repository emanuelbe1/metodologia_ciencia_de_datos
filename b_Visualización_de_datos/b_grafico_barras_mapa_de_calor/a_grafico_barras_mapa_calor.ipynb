Cada entrada muestra el tiempo de retraso promedio (en minutos) para una aerolínea y mes diferentes (todos en el año 2015). Las entradas negativas indican vuelos que (_en promedio_) tendían a llegar temprano. # Gráfico de barras

Queremos crear un grafico de barras que muestre el retraso promedio en llegada para los vuelos de Spirit Airlines (Código: **NK**) por mes.


```python
# Ancho y altura
plt.figure(figsize=(10,6))

# Titulo del gráfico
plt.title("Promedio de retraso en llegada de los vuelos de Spirit Airlines por mes")

# Retraso promedio de llegada para vuelos de Spirit Airlines por mes
sns.barplot(x=datos_vuelos.index, y=datos_vuelos['NK'])

plt.ylabel("Retraso en llegada (en minutos)")
plt.xlabel('Mes')
```

> Debemos seleccionar la columna `flight_data.index`. No es posible usar `flight_data['Month']` debido a que usamos la columna 'Month' para indexar las filas. 

 <h2>Mapa de calor(<i>Heatmap</i>)</h2>
<br>
Vamos a crear un mapa de calor para visualizar rápidamente patrones en `flight_data`. Cada celda está configurada con colores según el valor correspondiente.

```python
# Configuramos el ancho y el alto de la figura
plt.figure(figsize=(14,7))
plt.title("Promedio de retraso en llegada de los vuelos de cada aerolinea por mes")
sns.heatmap(data=datos_vuelos, annot=True)

plt.xlabel("Aerolinea")
plt.ylabel('Mes')
```

> `annot=True`  Esto asegura que los valores de cada celda aparezcan en el gráfico. Omitir este parametro elimina los números de cada una de las celdas.

 Si observamos detenidamente, los meses hacia el final del año (especialmente los meses 9 a 11) parecen relativamente oscuros para todas las aerolíneas. Esto sugiere que las aerolíneas son mejores (en promedio) para mantenerse a horario durante estos meses.
