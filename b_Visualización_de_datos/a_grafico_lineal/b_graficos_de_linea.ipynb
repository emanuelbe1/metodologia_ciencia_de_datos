# Graficando un subconjunto de los datos

Hemos trazado un grafico de línea para _todas_ las columnas del conjunto de datos. Ahora vamos a trazar un _subconjunto_ de las columnas.

Comenzamos imprimiendo los nombres de las columnas

```python
list(datos_spotify.columns)
```

Vamos a graficar dos lineas correspondientes a las dos primeras columnas del dataframe.

```python
plt.figure(figsize=(14,6)) #Configuramos el tamaño.
plt.title("Reproducciones diarias globales de canciones populares en 2017-2018") #Agregar titulo

# Gráfico de línea mostrando reproducciones globales diarias de 'Shape of You'
sns.lineplot(data=datos_spotify['Shape of You'], label="Shape of You")
#Gráfico de línea mostrando reproducciones globales diarias de 'Despacito'
sns.lineplot(data=datos_spotify['Despacito'], label="Despacito")
plt.xlabel("Fecha")
```
