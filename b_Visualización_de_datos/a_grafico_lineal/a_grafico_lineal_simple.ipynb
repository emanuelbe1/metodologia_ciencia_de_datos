Vamos a cargar un conjunto de datos con los rankings historicos de la FIFA para seis países: Argentina (ARG), Brasil (BRA), España (ESP), Francia (FRA), Alemania (GER), and Italia (ITA).

- especificamos la ubicación de los datos
- usamos la ruta de archivo para cargar el contenido de los datos. 

```python
# Leemos el conjunto de datos
datos_fifa = pd.read_csv('fifa.csv', index_col="Date", parse_dates=True)
```

```python
datos_fifa.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Grafico de linea mostrando cómo evolucionaron los rankings de fifa para Argentina a través del tiempo. 
# El eje X corresponde a los datos de la columna 'Date', el eje Y a datos de la columna 'ARG'
sns.lineplot(data=datos_fifa['ARG'])
```
