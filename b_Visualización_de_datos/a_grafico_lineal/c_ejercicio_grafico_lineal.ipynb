```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Cargamos los datos
<i>Parseamos</i> la columna 'Date' y la usamos cómo indice.

```python
datos_museo = pd.read_csv('museum_visitors.csv', index_col='Date', parse_dates=True)
```

```python
datos_museo.index.year.unique() #hay datos entre el 2014 y el 2018
```

## Gráficamos los datos

Creamos un grafico de lineas que muestre el numero de visitantes para cada museo en el tiempo. Nuestro gráfico va a tener cuatro lineas, una para cada museo.

```python
# Line chart showing the number of visitors to each museum over time
plt.figure(figsize=(14,6))
sns.lineplot(data=datos_museo)
```

## Gráfico para el museo Avila Adobe

Creamos un gráfico de lineas que muestra cómo evoluciono en el tiempo el número de visitantes para el museo Avila Adobe.

```python
datos_museo.columns
```

```python
# Line plot showing the number of visitors to Avila Adobe over time
sns.lineplot(data=datos_museo['Avila Adobe'])
plt.xlabel('Año')
plt.title('Visitas al múseo Avila Adobe')
plt.ylabel('Número de visitas')
```
